```python
# Calculate PCA matrix 

save = '../apollo_mean_car_shape'
beta = 10

from pytorch3d.io import load_obj, save_obj
import os 
import numpy as np
import torch


V = np.zeros((79,1352*3))
for car in range(79):
    vert, face, _ = load_obj(f'{car}.obj')
    V[car] = vert.view(-1).numpy()
    
# # compute mean shape
M = V.mean(0).reshape((1,1352*3))


from sklearn.decomposition import PCA

V_hat = V - M
# compute PCA matrix
pca = PCA(n_components=beta)
pca.fit(V_hat)
P = pca.components_

# carculate betas
betas = np.zeros((79,10))
for car in range(79):
    b = (V[car].reshape((1,1352*3)) - M) @ P.T
    betas[car] = b.reshape(beta)
    
# save results
np.save(os.path.join(save, 'P.npy'), P)
np.save(os.path.join(save, 'V.npy'), V)
np.save(os.path.join(save, 'beta.npy'), betas)
```

```python
# compute pca, betas per car part
save = '../apollo_mean_car_shape'
beta = 10

import os
import json
from pytorch3d.io import load_obj
import numpy as np
# Load part indices

with open('top_indices.json') as f:
    top_indices = json.load(f)

with open('body_indices.json') as f:
    body_indices = json.load(f)

with open('head_indices.json') as f:
    head_indices = json.load(f)
top_N, body_N, head_N = len(top_indices), len(body_indices), len(head_indices)
    
    
top_verts = {}
body_verts = {}
head_verts = {}
verts = {}

# Load V
top_V = np.zeros((79,top_N*3))
body_V = np.zeros((79,body_N*3))
head_V = np.zeros((79,head_N*3))
for car in range(79):
    vert, face, _ = load_obj(f'{car}.obj')

    top_V[car] = vert[top_indices].view(-1).numpy()
    body_V[car] = vert[body_indices].view(-1).numpy()
    head_V[car] = vert[head_indices].view(-1).numpy()

# # compute mean shape
top_M = top_V.mean(0, keepdims=True)
body_M = body_V.mean(0, keepdims=True)
head_M = head_V.mean(0, keepdims=True)

from sklearn.decomposition import PCA

top_V_hat = top_V - top_M
body_V_hat = body_V - body_M
head_V_hat = head_V - head_M

# compute PCA matrix
top_pca = PCA(n_components=beta)
body_pca = PCA(n_components=beta)
head_pca = PCA(n_components=beta)

top_pca.fit(top_V_hat)
head_pca.fit(head_V_hat)
body_pca.fit(body_V_hat)

top_P = top_pca.components_
head_P = head_pca.components_
body_P = body_pca.components_

# carculate betas
top_betas = np.zeros((79,10))
head_betas = np.zeros((79,10))
body_betas = np.zeros((79,10))
for car in range(79):
    top_b = (top_V[car].reshape((1,top_N*3)) - top_M) @ top_P.T
    head_b = (head_V[car].reshape((1,head_N*3)) - head_M) @ head_P.T
    body_b = (body_V[car].reshape((1,body_N*3)) - body_M) @ body_P.T
    
    top_betas[car] = top_b.reshape(beta)
    head_betas[car] = head_b.reshape(beta)
    body_betas[car] = body_b.reshape(beta)
    
# save results
np.save(os.path.join(save, 'top_M.npy'), top_M)
np.save(os.path.join(save, 'top_P.npy'), top_P)
np.save(os.path.join(save, 'top_V.npy'), top_V)
np.save(os.path.join(save, 'top_beta.npy'), top_betas)

np.save(os.path.join(save, 'head_M.npy'), head_M)
np.save(os.path.join(save, 'head_P.npy'), head_P)
np.save(os.path.join(save, 'head_V.npy'), head_V)
np.save(os.path.join(save, 'head_beta.npy'), head_betas)

np.save(os.path.join(save, 'body_M.npy'), body_M)
np.save(os.path.join(save, 'body_P.npy'), body_P)
np.save(os.path.join(save, 'body_V.npy'), body_V)
np.save(os.path.join(save, 'body_beta.npy'), body_betas)
```
